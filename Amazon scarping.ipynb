```python
# Importing Libraries
import pandas as pd
from bs4 import BeautifulSoup
import requests
```

```python
#send get request to the webpage server to get the source code of the page
url = "https://www.amazon.in/s?k=saree&ref=nb_sb_noss_2"
page11 = requests.get(url)

soup11 = BeautifulSoup(page11.content)
product_brand =[]
price = []    #empty list
rating = []   #empty list
img_url = []   #empty list


for i in soup11.find_all("span",class_="a-size-base-plus a-color-base"):
    product_brand.append(i.text)

for i in soup11.find_all("span",class_="a-price-whole"):
    price.append(i.text)
for i in soup11.find_all("span",class_="a-icon-alt"):
    rating.append(i.text)
    
for i in soup11.find_all("img",class_="s-image"):
    img_url.append(i.get("src"))
    

saree=pd.DataFrame({})
saree['product_brand']=product_brand[:50]
saree['price']=price[:50]
saree['Rating']=rating[:50]
saree['img_url'] = img_url[:50]
saree
```

```python
saree.to_excel('saree_data.xls', index=False)
```
